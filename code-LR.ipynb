```python
import pandas as pd
import numpy as np
data = pd.read_csv("HousingData.csv")
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data['CRIM'].fillna(data['CRIM'].mean(), inplace = True)
data['ZN'].fillna(data['ZN'].median(), inplace = True)
data['INDUS'].fillna(data['INDUS'].median(), inplace = True)
data['CHAS'].fillna(data['CHAS'].median(), inplace = True)
data['AGE'].fillna(data['AGE'].mean(), inplace = True)
data['LSTAT'].fillna(data['LSTAT'].median(), inplace = True)
```

```python
data['CRIM'].median()
```

```python
data.loc[(data['CRIM'].median() < data.CRIM), "COL"] = 1
data.loc[(data['CRIM'].median() > data.CRIM), "COL"] = 0
```

```python
data
```

```python
data.isnull().sum()
```

```python
x = data.drop('COL',axis = 1)
y = data['COL']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
x_train,x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 1)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_prediction = model.predict(x_test)
```

```python
import numpy as np
output = pd.DataFrame({'CRIM' : x_test['CRIM'], 'Prediction' : np.array(y_prediction)})
```

```python
output.to_csv('output.csv')
```
